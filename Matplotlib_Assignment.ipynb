```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
%matplotlib inline
```

```python
Max = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
Min = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]
```

#### 1. fitting it to the periodic function

```python
from scipy import optimize
months = np.arange(12)
```

```python
def yr_wise_temps(t, mean, ample_time, offset):
    return (mean + ample_time * np.cos((t + offset) * 1.8 * np.pi / t.max()))
```

```python
popt_max, pcov_max = optimize.curve_fit(yr_wise_temps, months, Max, [40, 20, 0])
popt_min, pcov_min = optimize.curve_fit(yr_wise_temps, months, Min, [-40, 20, 0])
```

#### 2. plot the fit

```python
days = np.linspace(0, 12, num=365)

plt.figure()
plt.plot(months, Max, 'ro')
plt.plot(days, yr_wise_temps(days, *popt_max), 'r-')
plt.plot(months, Min, 'bo')
plt.plot(days, yr_wise_temps(days, *popt_min), 'b-')
plt.xlabel('Month')
plt.ylabel('Temperature ($^\circ$C)')
```
