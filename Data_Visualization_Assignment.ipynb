```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
%matplotlib inline
```

```python
data = load_iris()
```

```python
data.data
```

```python
data.feature_names
```

```python
data.target_names
```

```python
df = pd.DataFrame(data.data, columns= data.feature_names)
df['target'] = data.target
```

```python
df
```

```python
pca = PCA(n_components= 3)
```

```python
transformed_data = pca.fit_transform(df[df.drop('target', axis= 1).columns].to_numpy())
transformed_data
```

```python
fig= plt.figure(figsize= (12, 10))
ax = Axes3D(fig)
ax.scatter(xs= transformed_data[:, 0], ys= transformed_data[:, 1], zs= transformed_data[:, 2], c= df.target)
ax.set_xlabel('X - axis')
ax.set_ylabel('Y - axis')
ax.set_zlabel('Z - axis')
```
